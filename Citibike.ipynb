```python
import pandas as pd
```

```python
junedata = pd.read_csv("202006-citibike-tripdata.csv/202006-citibike-tripdata.csv")
maydata = pd.read_csv("202005-citibike-tripdata.csv/202005-citibike-tripdata.csv")
aprildata = pd.read_csv("202004-citibike-tripdata.csv/202004-citibike-tripdata.csv")
```

```python
print (maydata.head(5))
```

```python
junedf = pd.DataFrame(data = junedata)
maydf = pd.DataFrame(data = maydata)
aprildf = pd.DataFrame(data = aprildata)
```

```python
trip_df = junedf.append(maydf, ignore_index=True)
```

```python
trip_df = trip_df.append(aprildf, ignore_index=True)
```
